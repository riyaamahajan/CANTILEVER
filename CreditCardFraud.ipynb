```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data=pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
#checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transaction & fraudulent transaction
credit_card_data['Class'].value_counts()
```

```python
# 0-->>normal transaction 1->> fraudlent transaction
```

```python
#separate data for transaction
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
#under sampling
# to build a sample dataset containing similar distribution of normal transaction and fraundulent Transaction
```

```python
legit_sample=legit.sample(n=239)
```

Concatenating two data frames

```python
new_data=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

Splitting data into Features and Targets

```python
X=new_data.drop(columns='Class',axis=1)
Y=new_data['Class']
```

```python
print(X
      )
```

```python
print(Y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,x_train.shape,x_test.shape)
```

MODEL TRAINING
Logistic Regression

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

MODEL VALUATION


```python
#accuracy on training data
X_train_prediction=model.predict(x_train)
train_data_accuracy=accuracy_score(X_train_prediction,y_train)
```

```python
print('Accuracy of training data:',train_data_accuracy)
```

```python
X_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(X_test_prediction,y_test)
```

```python
print('Accuracy of training data:',test_data_accuracy)
```
